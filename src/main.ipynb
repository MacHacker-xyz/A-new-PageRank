```python
import utils
import numpy as np

# 选择阻尼系数和迭代次数
alpha,N = 0.85,100
# 导入节点分类的标记值
target = utils.target_reader("../input/cora_target.csv")
# 初始化个性化概率向量v
v = np.array(target,dtype=np.float64)
# 设置个个性化向量比例
v[target==0],v[target==1],v[target==2],v[target==3],v[target==4],v[target==5],v[target==6] = 1.6,1.4,1.2,1.0,0.8,0.6,0.4
# 计算的出个性化向量的最终表达式
v=np.reshape(v/v.sum(),(1,2708))
# 导入网络有向图
P = utils.graph_reader("../input/cora_edges.csv")
# 标准化网络连接矩阵
P = utils.normalize_adjacency_matrix(P*v)
# 计算传态矩阵
E = np.matmul(np.ones((P.shape[0],1)),v)
# 计算转移矩阵
P = alpha*P+(1-alpha)*E
# 初始化PagerRank向量
pi = np.ones((1,P.shape[0]))/P.shape[0]
# 迭代固定次数得到PagerRank向量
for i in range(1000):
    pi = np.matmul(pi,P)
pi = pi[0]
result3 = pi
```
